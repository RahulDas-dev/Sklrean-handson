### Enabling Basic Settings

```python
pd.set_option("display.max_colwidth", 1000)
```

### Dataset Path validation

```python
parent_directory = os.path.abspath(
    ".."
)  # Should Point to root directory /Project Directory
os.chdir(parent_directory)

datset_dir = os.path.join(parent_directory, "data")

if os.path.isdir(datset_dir) is False:
    raise NotADirectoryError(f"Directory {datset_dir} not valid")
else:
    print(f"datset_dir {datset_dir}")

dataset_path = os.path.join(datset_dir, "zomato.csv")


if os.path.isfile(dataset_path) is False:
    raise NotADirectoryError(f"dataset_path {dataset_path} not valid")
else:
    print(f"datset_path {dataset_path}")
```

### Dataset Loading

```python
dataset = pd.read_csv(dataset_path, encoding="ISO-8859-1")

print(f"dataset {dataset.shape}")
dataset.head()
```

### Shuffling the Dataset

```python
dataset = dataset.sample(frac=1, random_state=32).reset_index(drop=True)

print(f"dataset {dataset.shape}")
dataset.head()
```

### Spliting The Data into Test & Train

```python
total_size = dataset.shape[0]

train_size = math.ceil(total_size * 0.8)

test_size = total_size - train_size

print(f"total_size {total_size}, train_size {train_size}, test_size {test_size}")

train_df = dataset[:train_size].copy()
test_df = dataset[train_size:].copy()


train_df.to_csv(os.path.join(datset_dir, "train_data.csv"), index=False)
test_df.to_csv(os.path.join(datset_dir, "test_data.csv"), index=False)
```
